# Scraping a table from Wikipedia

```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
```

```python
with urlopen('https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(PPP)') as response:
    soup = BeautifulSoup(response, 'html.parser')
table = soup.find("table", {"class":"wikitable"})
table_body = table.find('tbody')
rows = table_body.find_all('tr')
i=1
for row in rows:
    print(f"Row number: {i}")
    print(row.text)
    print('------------')
    i+=1
```

```python
for row in rows[4:198]:
    print(row.text)
    print('------------')
```

```python
i=1
for row in rows[4:198]:
    print(f"Row number: {i}")
    cells = row.find_all('td')
    for cell in cells:
        tekst = cell.text.strip()
        tekst.replace(" ", "")
        print(tekst)
        print('++++++++')
    print('------------')
```

```python
rank_IMF = [row.find_all('td')[0].text.strip() for row in rows[4:197]]
```

```python
print(rank_IMF)
```

```python
country_IMF = [row.find_all('td')[1].text.strip() for row in rows[4:197]]
```

```python
print(country_IMF)
```

```python
gdp_IMF = [row.find_all('td')[2].text.strip() for row in rows[4:197]]
```

```python
print(gdp_IMF)
```

```python
import pandas as pd
```

```python
table = {'Rank' : rank_IMF,
         'Country/Territory' : country_IMF,
         'GDP(millions of $)' : gdp_IMF}
```

```python
list_by_IMF = pd.DataFrame(table, columns = ['Rank', 'Country/Territory', 'GDP(millions of $)'])
```

```python
print(list_by_IMF)
```
